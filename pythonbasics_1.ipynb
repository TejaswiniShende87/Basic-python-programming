```python
# Common Escape Characters 
```

```python
# Escape characters are special symbols (starting with \) used to add formatting or special meaning inside a string.
```

```python
# New Line → \n
# Moves text to the next line
```

```python
print("Hello\nWorld")
```

```python
# Tab Space → \t
# Adds horizontal space (tab)
```

```python
print("Hello\tWorld")
```

```python
# Backslash → \\
```

```python
print("C:\\Users\\Rahul")
```

```python
# Single Quote → \'
# Used when string is inside single quotes
```

```python
print('It\'s Python')
```

```python
# Double Quote → \"
# Used when string is inside double quotes
```

```python
print("He said \"Hello\"")
```

```python
# Backspace → \b
# Removes one character before it
```

```python
print("Helloo\b")
```
