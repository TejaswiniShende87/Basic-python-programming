```python
# 17. Write a Python program to check whether three sides form a valid triangle, identify its type, and calculate its perimeter.
```

```python
a = float(input("Enter first side: "))
b = float(input("Enter second side: "))
c = float(input("Enter third side: "))
```

```python
if (a + b > c) and (a + c > b) and (b + c > a):
    print("It is a valid triangle.")

    # Find perimeter
    perimeter = a + b + c
    print("Perimeter of triangle:", perimeter)

    # Identify type of triangle
    if a == b == c:
        print("It is an Equilateral triangle.")
    elif a == b or b == c or a == c:
        print("It is an Isosceles triangle.")
    else:
        print("It is a Scalene triangle.")

else:
    print("The given sides do not form a valid triangle.")
```

```python
# 22. Write a Python program to accept a password, check its length, validate whether it contains letters and digits, and display password strength.
```

```python
password = input("Enter your password: ")
```

```python
has_letter = False
has_digit = False
```

```python
for ch in password:
    if ch.isalpha():
        has_letter = True
    elif ch.isdigit():
        has_digit = True
```

```python
length = len(password)
```

```python
if length < 6:
    print("Weak password (too short).")

elif has_letter and has_digit and length >= 8:
    print("Strong password.")

elif has_letter and has_digit:
    print("Moderate password.")

else:
    print("Weak password (must contain both letters and digits).")
```
