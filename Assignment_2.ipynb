```python
# 3. Write a Python program to check whether a number is prime using a function, count its factors, and display the result.
```

```python
def check_prime(n):
    factor_count = 0

    for i in range(1, n + 1):
        if n % i == 0:
            factor_count += 1

    if factor_count == 2:
        return factor_count, True
    else:
        return factor_count, False
```

```python
num = int(input("Enter a number: "))
```

```python
factors, is_prime = check_prime(num)
```

```python
print("Number of factors:", factors)
```

```python
if is_prime:
    print(num, "is a Prime Number.")
else:
    print(num, "is Not a Prime Number.")
```

```python
# 8. Write a Python program to calculate the factorial using recursion, count recursive calls, and display the result.
```

```python
call_count = 0
```

```python
def factorial(n):
    global call_count
    call_count += 1

    if n == 0 or n == 1:
        return 1
    else:
        return n * factorial(n - 1)
```

```python
num = int(input("Enter a non-negative integer: "))
```

```python
if num < 0:
    print("Factorial is not defined for negative numbers.")
else:
    result = factorial(num)

    print("Factorial of", num, "is", result)
    print("Number of recursive calls =", call_count)
```

```python
# 9. Write a Python program to calculate the sum of digits using recursion, reverse the number, and count its digits.
```

```python
def sum_digits(n):
    if n == 0:
        return 0
    return (n % 10) + sum_digits(n // 10)
```

```python
def reverse_number(n, rev=0):
    if n == 0:
        return rev
    return reverse_number(n // 10, rev * 10 + n % 10)
```

```python
def count_digits(n):
    if n == 0:
        return 0
    return 1 + count_digits(n // 10)
```

```python
num = int(input("Enter a number: "))
```

```python
digit_count = 1 if num == 0 else count_digits(abs(num))
```

```python
print("Sum of digits =", sum_digits(abs(num)))
print("Reversed number =", reverse_number(abs(num)))
print("Number of digits =", digit_count)
```

```python
def divide_numbers(a, b):
    try:
        result = a / b
        print(f"Result: {a} / {b} = {result}")
    except ZeroDivisionError:
        print("Error: Division by zero is not allowed.")
```

```python
numbers = [(10, 2), (5, 0), (8, 4), (9, 3)]
```

```python
for num in numbers:
    divide_numbers(num[0], num[1])

print("\nProgram execution completed successfully.")
```

```python
# 21. Write a Python program to handle multiple exceptions, display the exception type, and log the error message.
```

```python
import logging
```

```python
logging.basicConfig(
    filename="error_log.txt",
    level=logging.ERROR,
    format="%(asctime)s - %(levelname)s - %(message)s"
)
```

```python
def process_data(a, b):
    try:
        result = a / b
        print("Result:", result)

    except ZeroDivisionError as e:
        print("Error Type:", type(e).__name__)
        print("Message:", e)
        logging.error(f"ZeroDivisionError: {e}")

    except TypeError as e:
        print("Error Type:", type(e).__name__)
        print("Message:", e)
        logging.error(f"TypeError: {e}")

    except Exception as e:
        print("Error Type:", type(e).__name__)
        print("Message:", e)
        logging.error(f"Unexpected Error: {e}")
```

```python
test_cases = [
    (10, 2),
    (5, 0),
    (10, "a"),
]
```

```python
for a, b in test_cases:
    process_data(a, b)

print("\nProgram completed. Check 'error_log.txt' for logged errors.")
```

```python
# 22. Write a Python program to demonstrate try, except, and finally, and display execution flow messages.
```

```python
def divide(a, b):
    print("Inside function: Starting division process")

    try:
        print("Try block: Attempting division")
        result = a / b
        print("Result =", result)

    except ZeroDivisionError:
        print("Except block: Cannot divide by zero!")

    finally:
        print("Finally block: Execution completed (always runs)")

    print("Function ended\n")
```

```python
print("Case 1:")
divide(10, 2)

print("Case 2:")
divide(5, 0)

print("Program execution finished.")
```

```python
# 23. Write a Python program to raise a custom exception, handle the exception, and display a user-friendly message.
```

```python
class InvalidAgeError(Exception):
    pass
```

```python
def check_age(age):
    try:
        if age < 18:
            raise InvalidAgeError("Age must be 18 or above to proceed.")
        else:
            print("Access granted. You are eligible.")

    except InvalidAgeError as e:
        print("Custom Exception Caught!")
        print("Message:", e)
```

```python
check_age(20)
print()
check_age(15)
```

```python
# 24. Write a Python program to handle a file-not-found exception, create the missing file, and write default content into it.
```

```python
filename = "data.txt"
```

```python
try:

    with open(filename, "r") as file:
        content = file.read()
        print("File Contents:\n", content)

except FileNotFoundError:
    print("File not found. Creating a new file with default content...")


    with open(filename, "w") as file:
        file.write("Default File Created\n")
        file.write("This is sample default content.\n")
        file.write("File was missing and now created automatically.\n")

    print("File created successfully with default content.")

finally:
    print("Program execution completed.")
```
