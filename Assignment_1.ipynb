```python
# 6. Write a Python program to calculate the factorial of a number, count its digits, and display the reverse of the number.
```

```python
number = int(input("Enter the number"))
```

```python
factorial = 1
for i in range(1, num + 1):
    factorial *= i

digit_count = len(str(abs(num)))

reverse_num = int(str(abs(num))[::-1])

if num < 0:
    reverse_num = -reverse_num

print("Factorial =", factorial)
print("Number of digits =", digit_count)
print("Reverse of the number =", reverse_num)
```

```python
# 7. Write a Python program to generate the Fibonacci series, calculate its sum, and display the largest term generated.
```

```python
n = int(input("Enter the number of terms: "))
a, b = 0, 1
fib_series = []
sum_series = 0
for i in range(n):
    fib_series.append(a)
    sum_series += a
    a, b = b, a + b
print("Fibonacci Series:", fib_series)
print("Sum of the series:", sum_series)

if fib_series:
    print("Largest term generated:", max(fib_series))
else:
    print("No terms generated.")
```

```python
# 11. Write a Python program to reverse a string, count the vowels, and count the consonants.
```

```python
text = input("Enter a string: ")
reversed_text = text[::-1]

vowels = 0
consonants = 0
for char in text.lower():
    if char.isalpha():
        if char in "aeiou":
            vowels += 1
        else:
            consonants += 1

print("Reversed String:", reversed_text)
print("Number of Vowels:", vowels)
print("Number of Consonants:", consonants)
```
